```python
import string

contractions = {
    "i'm": "i am",
    "i've": "i have",
    "i'll": "i will",
    "i'd": "i would",
    "you're": "you are",
    "you've": "you have",
    "you'll": "you will",
    "he's": "he is",
    "she's": "she is",
    "it's": "it is",
    "we're": "we are",
    "they're": "they are",
    "don't": "do not",
    "doesn't": "does not",
    "didn't": "did not",
    "can't": "cannot",
    "won't": "will not",
    "isn't": "is not",
    "aren't": "are not",
    "wasn't": "was not",
    "weren't": "were not",
    "haven't": "have not",
    "hasn't": "has not",
    "wouldn't": "would not",
    "couldn't": "could not",
    "shouldn't": "should not",
    "that's": "that is",
    "what's": "what is",
    "there's": "there is"
}
num_to_word = {
    "0": "zero", "1": "one", "2": "two", "3": "three",
    "4": "four", "5": "five", "6": "six", "7": "seven",
    "8": "eight", "9": "nine", "10": "ten"
}

def expand_contractions(text):
    words = text.split()
    result = []
    for word in words:
        clean_word = word.strip(string.punctuation)
        if clean_word in contractions:
            word = word.replace(clean_word, contractions[clean_word])
        result.append(word)
    return " ".join(result)

def convert_numbers(text):
    words = text.split()
    result = []
    for word in words:
        if word in num_to_word:
            result.append(num_to_word[word])
        else:
            result.append(word)
    return " ".join(result)

def normalize(text):
    text = text.lower()
    
    text = expand_contractions(text)
    text = convert_numbers(text)
    result = ""
    for char in text:
        if char not in string.punctuation:
            result += char
    text = result

    words = text.split()
    text = " ".join(words)
    return text
with open("/kaggle/input/datasets/nourtlm/222222222222222222222/Task2.txt", "r") as f:
    text = f.read()

print("Avant :", text)
print("Après :", normalize(text))
```

```python

n = {
    "0": "zero", "1": "one", "2": "two", "3": "three", "4": "four",
    "5": "five", "6": "six", "7": "seven", "8": "eight", "9": "nine", "10": "ten"
}
words_numbr = {
    "one": "one", "two": "two", "three": "three", "four": "four",
    "five": "five", "six": "six", "seven": "seven", "eight": "eight",
    "nine": "nine", "ten": "ten"
}
def normalisation(texte):

    texte = texte.lower()

    mots = texte.split()

    mots_normalises = []
    for w in mots:

        if w.isdigit() and w in n:
            mots_normalises.append(n[w])

        elif w in words_numbr:
            mots_normalises.append(words_numbr[w])

        else:
            new_w = ""
            for char in w:
                if char.isalpha() or char == " ":
                    new_w += char
            if new_w != "":
                mots_normalises.append(new_w)

    texte = " ".join(mots_normalises)

    return texte


docs = [
    "Today she cooked 4 bourak. Later, she added two chamiyya and 1 pizza.",
    "Five pizza were ready, but 3 bourak burned!",
    "We only had 8 chamiyya, no pizza, and one tea.",
    "Is 6 too much? I ate nine bourak lol."
]

for i, doc in enumerate(docs, 1):
    print(f"D{i} original  : {doc}")
    print(f"D{i} normalise : {normalisation(doc)}")
    print()
```
